```python
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
geoapify_key = "819f0915f55842cfa604558744bda981"
```

```python
filepath = '../DATA_SET/cleaned_data.csv'
df = pd.read_csv(filepath)
df.head(10)
```

```python
Location_dir = {'1':'Portugal',
                '2':'Germany',
                '3':'Spain',
                '4':'Italy',
                '5':'Netherlands',
                '6':'Britain',
                '7':'Lithuania',
                '8':'Angola',
                '9':'Cabo Verde',
                '10':'Guinea',
                '11':'Mozambique',
                '12':'Sao Tome e Principe',
                '13':'Turkey',
                '14':'Brazil',
                '15':'Romania',
                '16':'Moldova (Republic of)',
                '17':'Mexico',
                '18':'Ukrainia',
                '19':'Russia',
                '20':'Cuba',
                '21':'Colombia'}
```

```python
location_df = pd.DataFrame({'Nationality':range(1,22),
                           'Country':['Portugal','Germany','Spain','Italy','Netherlands','Britain','Lithuania',
                                      'Angola','Cabo Verde','Guinea','Mozambique','Sao Tome e Principe','Turkey','Brazil',
                                      'Romania','Moldova (Republic of)','Mexico','Ukrainia','Russia','Cuba','Colombia']})
location_df.head()
```

```python
base_url = 'https://api.geoapify.com/v1/geocode/search?'
lat = []
lng = []
for i in location_df['Country']:
    params = {'apiKey':geoapify_key,
             'country': i} 
    response = requests.get(base_url,params = params).json()
    try:
        lat.append(response['features'][0]['geometry']['coordinates'][1])
        lng.append(response['features'][0]['geometry']['coordinates'][0])
        print(f'{i} is found')
    except:
        print(f'{i} is not found！！！！！')
        lat.append(0)
        lng.append(0)
location_df['Latitude'] = lat
location_df['Longitude'] = lng
```

```python
location_df.head(10)
```

```python
new_df = pd.merge(df,location_df,on = 'Nationality',how='left')
new_df.head(20)
```

```python
new_df.to_csv('../DATA_SET/CleanedDataWithLocation.csv',header=True,index = False)
```
